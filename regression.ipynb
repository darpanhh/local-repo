```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


x = np.array([1, 2, 3, 4, 5])
y = np.array([1.2, 1.9, 3.2, 3.8, 5.1])


def exponential_func(x, a, b):
    return a * np.exp(b * x)


params, covariance = curve_fit(exponential_func, x, y, p0=(1, 0.1))  # initial guesses for a and b

x_fit = np.linspace(min(x), max(x), 100)
y_fit = exponential_func(x_fit, *params)

plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_fit, y_fit, color='green', label=f'Fitted Exponential Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Least Squares Exponential Curve Fitting')
plt.grid(True)
plt.show()


print(f"Fitted parameters: a = {params[0]:.4f}, b = {params[1]:.4f}")


```

Cramer's rule

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([1.2, 1.9, 3.2, 3.8, 5.1], dtype=float)
Y = np.log(y)
n = len(x)
sum_x = np.sum(x)
sum_Y = np.sum(Y)
sum_x2 = np.sum(x * x)
sum_xY = np.sum(x * Y)
D = n * sum_x2 - sum_x * sum_x
D_A = sum_Y * sum_x2 - sum_x * sum_xY
D_b = n * sum_xY - sum_Y * sum_x


A = D_A / D
b = D_b / D
a = math.exp(A)

print(f"Fitted parameters using Cramer's Rule:\na = {a:.4f}, b = {b:.4f}")
print(f"Exponential model: y = {a:.4f} * e^({b:.4f}x)")


x_fit = np.linspace(min(x), max(x), 100)
y_fit = a * np.exp(b * x_fit)

plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_fit, y_fit, color='green', label='Fitted Exponential Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(rf'Exponential fit using Cramer\'s Rule: $y ={a:.2f}\cdot e^{{{b:.2f}x}}$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 2, 3, 4, 5,6,7,8])
y = np.array([2, 5, 9, 3, 11,12,14,16])

N = len(x)


sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xx = np.sum(x * x)
sum_xy = np.sum(x * y)


m = (N * sum_xy - sum_x * sum_y) / (N * sum_xx - sum_x**2)
c = (sum_y - m * sum_x) / N

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

y_fit = m * x + c


plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_fit, color='red', label='Fitted Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Curve Fitting Using Least Squares Method')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Step 1: Generate clean exponential data
x = np.array([1, 2, 3, 4, 5], dtype=float)
true_a = 1.0
true_b = 0.5
y_clean = true_a * np.exp(true_b * x)

# Step 2: Add random noise to y values
noise = np.random.normal(0, 0.2, size=x.shape)  # Mean=0, Std=0.2
y = y_clean + noise  # noisy data

# Step 3: Convert y to Y = ln(y)
# Important: Remove or filter out points where y <= 0 (log undefined)
valid_indices = y > 0
x = x[valid_indices]
y = y[valid_indices]
Y = np.log(y)

# Step 4: Compute necessary sums
n = len(x)
sum_x = np.sum(x)
sum_Y = np.sum(Y)
sum_x2 = np.sum(x * x)
sum_xY = np.sum(x * Y)

# Step 5: Solve using Cramer's Rule
D = n * sum_x2 - sum_x * sum_x
D_A = sum_Y * sum_x2 - sum_x * sum_xY
D_b = n * sum_xY - sum_Y * sum_x

A = D_A / D
b = D_b / D
a = math.exp(A)

print(f"Fitted parameters using Cramer's Rule (with noise):\na = {a:.4f}, b = {b:.4f}")
print(f"Fitted model: y = {a:.4f} * e^({b:.4f}x)")

# Step 6: Plot
x_fit = np.linspace(min(x), max(x), 100)
y_fit = a * np.exp(b * x_fit)

plt.scatter(x, y, color='red', label='Noisy Data Points')
plt.plot(x_fit, y_fit, color='green', label='Fitted Exponential Curve')
plt.plot(x, y_clean[valid_indices], 'bo--', label='True Curve (No Noise)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Exponential Fitting with Noise using Cramer\'s Rule')
plt.legend()
plt.grid(True)
plt.show()
```
